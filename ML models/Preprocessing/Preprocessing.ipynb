```python
import numpy as np
import cv2
import pandas as pd
from matplotlib import pyplot as plt


def subtract_gaussian_blur(img):
    gb_img = cv2.GaussianBlur(img, (0, 0), 25)
    return cv2.addWeighted(img, 4, gb_img, -4,128)

img = cv2.imread('13_left.jpeg',1)
img2=cv2.imread('13_left.jpeg',0)
#gb_img = cv2.GaussianBlur(img, (0, 0), 5)
img = subtract_gaussian_blur(img)


print(img2.shape)

height=int((img2.shape[0])/2)
width=(img2.shape[1])

print(width)

back=img2[0][0]

for i in range(width):
    if(img2[height][i]-back>=10):
        dist=i
        break
        
print(dist)

radius=int(width/2-dist)
print("R",radius)

cv2.circle(img,(width//2,height),95*radius//100, (0,0,0), 20,)



cv2.resize

rx=radius
ry=radius
r=256
resize_scale = r / max(rx, ry)
w = min(int(rx * resize_scale * 2), radius * 2)
h = min(int(ry * resize_scale * 2), radius * 2)
        
img = cv2.resize(img, (0,0), fx=resize_scale, fy=resize_scale)

plt.figure(100)
plt.imshow(img)
plt.show()


print(img.shape)

cv2.imwrite('preproc.png',img)
```
